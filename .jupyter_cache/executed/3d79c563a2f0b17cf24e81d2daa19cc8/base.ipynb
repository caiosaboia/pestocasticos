```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def initialize_lattice(n):
  return np.random.choice([0,1], size=(n,n))

def energy(lattice):
    """Calcula a energia total do sistema conforme o modelo de Ising."""
    E = 0
    n = lattice.shape[0]
    for i in range(n):
        for j in range(n):
            S = 2 * lattice[i, j] - 1  # Converte 0/1 para -1/+1
            neighbors = (2 * lattice[(i+1) % n, j] - 1) + (2 * lattice[i, (j+1) % n] - 1) + \
                        (2 * lattice[(i-1) % n, j] - 1) + (2 * lattice[i, (j-1) % n] - 1)
            E += -S * neighbors
    return E / 2  # Para evitar contagem dupla

def metropolis_step(lattice, beta):
    """Executa um passo do algoritmo de Metropolis-Hastings."""
    n = lattice.shape[0]
    for _ in range(n * n):  # n^2 tentativas de atualização
        i, j = np.random.randint(0, n, size=2)  # Escolhe um spin aleatório
        S = 2 * lattice[i, j] - 1  # Converte 0/1 para -1/+1
        neighbors = (2 * lattice[(i+1) % n, j] - 1) + (2 * lattice[i, (j+1) % n] - 1) + \
                    (2 * lattice[(i-1) % n, j] - 1) + (2 * lattice[i, (j-1) % n] - 1)
        dE = 2 * S * neighbors  # Mudança na energia
        if dE < 0 or np.random.rand() < np.exp(-beta * dE):
            lattice[i, j] = 1 - lattice[i, j]  # Inverte entre 0 e 1
    return lattice

def simulate_ising(n, beta_values, steps=10000):
    """Simula o modelo de Ising para diferentes valores de beta."""
    lattices = []
    for beta in beta_values:
        lattice = initialize_lattice(n)
        for _ in range(steps):
            lattice = metropolis_step(lattice, beta)
        lattices.append((beta, lattice))
    return lattices

def plot_lattices(lattices):
    """Plota as configurações finais para diferentes valores de beta."""
    fig, axes = plt.subplots(1, len(lattices), figsize=(12, 4))
    for ax, (beta, lattice) in zip(axes, lattices):
        ax.imshow(lattice, cmap='gray')
        ax.set_title(f'Beta = {beta:.2f}')
        ax.axis('off')
    plt.show()

def plot_lattice_table(lattice):
    """Plota a matriz em forma de tabela colorida."""
    plt.figure(figsize=(6,6))
    sns.heatmap(lattice, cmap="coolwarm", cbar=False, linewidths=0.5, linecolor='black', square=True)
    plt.title("Configuração da Matriz de Spins")
    plt.show()
```

```python
# Parâmetros da simulação
n = 5  # Tamanho da grade
beta_values = [0.5, 1.5, 2.5]  # Três valores de beta no intervalo (0,3)
lattices = simulate_ising(n, beta_values)
plot_lattices(lattices)

# Exibir uma das matrizes em forma de tabela colorida
plot_lattice_table(lattices[0][1])
```
